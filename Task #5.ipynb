**Basic Data Visualization**

```python
df.drop(columns = ['Country', 'Postal Code', 'City', 'Ship Mode'], axis = 1, inplace = True)
sns.pairplot(df,hue='Region')
```

```python
df=pd.read_csv('SampleSuperstore.csv')
for i in df[['Ship Mode', 'Segment', 'Region', 'Category']]:
  plt.figure(figsize=(12,5))
  sns.countplot(df[i])
  plt.show()
```

1)The Standard Shippping Mode is the most preffered Shipping Mode and Same Day Mode is the least preffered

2)Consumers use the highest no of shipping while the Home offices use the least

3)Highest sales are recorded in the West region while lowest are recorded in the Southern region

4)Office supplies are the most preffered products while Technological products are least preffered

```python
plt.figure(figsize=(16,6))
sns.countplot(x=df['State'])
plt.xticks(rotation='90')
```

```python
df.groupby(['Ship Mode'])['Sales','Profit'].sum().plot.bar(figsize=(12,6),
          title=' Ship Mode vs Sales and Profits')
```

Highest Sales and profits were recorded when the Standard Shipping Mode was used

```python
df.groupby(['State'])['Sales','Profit'].sum().plot.bar(figsize=(12,6),
          title=' State vs Sales and Profits')
```

California has the largest number of Sales and profits

```python
df.groupby(['Region'])['Sales','Profit'].sum().plot.bar(figsize=(12,6),
          title='Region vs Sales and Profits')
```

West region recorded the maximum sales and Profit

```python
df.groupby(['Segment'])['Sales','Profit'].sum().plot.bar(figsize=(12,6),
          title=' Segment wise Sales and Profits')
```

Consumers accounted to maximum sales and profits

```python
df.groupby(['Category'])['Sales','Profit'].sum().plot.bar(figsize=(12,6),
          title=' State vs Sales and Profits')
```

Maximum sales and profits were recorded for Technological supplies

```python
df.groupby(['Sub-Category'])['Sales','Profit'].sum().plot.bar(figsize=(12,6),
          title=' State vs Sales and Profits')
```

Maximum Sale of Phones and Chairs took place. Maximum profit was recorded for Copiers while sale of tables and Bookcases lead to a loss.

```python
df.groupby(['Category'])['Discount'].sum().plot.bar(figsize=(12,6),
          title=' Category wise discount')
```

Maximum Discount was given for Office supplies

```python
df.groupby(['Sub-Category'])['Discount'].sum().plot.bar(figsize=(12,6),
          title=' Sub-Category wise discount')
```

Maximum Discount was given for Binders

**Some complex analysis**

```python
df.groupby(['Region','Category'])['Sales','Profit'].sum().plot.bar(figsize=(12,6),
          title='Region vs Sales and Profits')
```

Sale of furniture in central region led to a Loss

**Pie Charts**

```python
df_region = df.groupby('Region')['Quantity'].sum().reset_index()
plt.figure(figsize = (12,8))
plt.pie(df_region['Quantity'], labels = df_region['Region'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Quantities ordered by each region', fontsize = 15)
```

```python
df_region = df.groupby('Category')['Quantity'].sum().reset_index()
plt.figure(figsize = (12,8))
plt.pie(df_region['Quantity'], labels = df_region['Category'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Quantities ordered for each category', fontsize = 15)
```

```python
df_region = df.groupby('Sub-Category')['Quantity'].sum().reset_index()
plt.figure(figsize = (12,8))
plt.pie(df_region['Quantity'], labels = df_region['Sub-Category'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Quantities ordered by each region', fontsize = 15)
```
